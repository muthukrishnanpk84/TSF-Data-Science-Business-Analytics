### Let's do some Visualization using the pie plot

```python
plt.figure(figsize=(12,6))
plt.pie(superstore['Category'].value_counts(), labels=superstore['Category'].value_counts().index, startangle=180, radius=1)
plt.title('Categories of Items for Sale')
plt.show()
```

```python
subgroup=superstore.groupby('Ship Mode')['Sales'].agg('count')
plt.figure(figsize=(12,6))
labels=superstore['Ship Mode'].value_counts().index
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice
plt.title('Shipping mode for Sale',color='black',fontsize=20)
plt.pie(x=subgroup,colors=colors,labels=labels,explode=explode,autopct='%1.0f%%', shadow=True, startangle=120,radius=1.2)
plt.show()
```

```python
subgroup1=superstore.groupby('Segment')['Sales'].agg('count')
plt.figure(figsize=(12,6))
labels=superstore['Segment'].value_counts().index
colors = ['gold', 'yellowgreen', 'lightcoral']
textprops = {"fontsize":15}
explode = (0.1, 0, 0)  # explode 1st slice
plt.title('Sales with Segment',color='black',fontsize=20,loc="left",)
plt.pie(x=subgroup1,colors=colors,labels=labels,explode=explode,autopct='%1.0f%%', shadow=True, startangle=110,radius=1.2,textprops=textprops)
plt.show()
```

```python
subgroup2=superstore.groupby('Region')['Sales'].agg('count')
plt.figure(figsize=(12,6))
labels=superstore['Region'].value_counts().index
colors = ['gold', 'yellowgreen', 'lightcoral','lightskyblue']
textprops = {"fontsize":15}
explode = (0.1, 0, 0,0)  # explode 1st slice
plt.title('Sales with region',color='black',fontsize=20,loc="left",)
plt.pie(x=subgroup2,colors=colors,labels=labels,explode=explode,autopct='%1.0f%%', shadow=True, startangle=110,radius=1.2,textprops=textprops)
plt.show()
```

```python
subgroup3=superstore.groupby('Category')['Sales'].agg('count')
plt.figure(figsize=(12,6))
labels=superstore['Category'].value_counts().index
colors = ['gold', 'yellowgreen', 'lightcoral']
textprops = {"fontsize":15}
explode = (0.1, 0, 0)  # explode 1st slice
plt.title('Sales with respect to the Category',color='black',fontsize=20,loc="right",)
plt.pie(x=subgroup3,colors=colors,labels=labels,explode=explode,autopct='%1.0f%%', shadow=True, startangle=90,radius=1.2,textprops=textprops)
plt.show()
```
